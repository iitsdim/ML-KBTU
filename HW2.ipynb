## Task 1.3 (3 points)

Let $X_1, \ldots, X_n$ is an i.i.d. sample from $\mathcal N(0, \sigma^2)$ and $\widehat \sigma = \frac 1n \sum\limits_{i=1}^n |X_i|$.

* Find bias, se and MSE of the estimation $\widehat \sigma$. Is this estimation unbiased? 
* If $\widehat \sigma$ turned out to be biased, fix it to get an unbiased estimation of $\sigma$. Find se of this corrected estimation. Is it consistent?

### YOUR SOLUTION HERE

As $ X_{1}, ..., X_{n} $ is an i.i.d (independent and identically distributed) sample, then it has certain properties, such as:

1. $ X_{1}, ..., X*{n} $ are independent
2. $ X_{i} $ has the same marginal distribution with CDF F, in other words, it has the same probability distribution as others.

$ \boldsymbol{(a)} $ Bias of standard deviation $ \widehat \sigma $ is the difference between expected value of std and actual value

$$
    bias(\widehat \sigma) = E[\widehat \sigma] - \sigma
$$

In order to find the expectation of sample standard deviation $ E[\widehat \sigma] $, firstly we need to calculate the expectation of gaussian distribution $ \mathcal{N}(0, \sigma^2) $. Guassian distribution is a continuous distribution, and the expectation of continuous random variable is

$$
    E(x) = \int\limits_{-\infty}^{+\infty} x f(x)dx =
$$

The probability density function of gaussian distribution of $ \mathcal{N}(\mu, \sigma^2) $ is

$$
    f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{\frac{(x-\mu)^2}{2\sigma^2}}
$$

Expectation $ E[X] $ of $ \mathcal{0}(\mu, \sigma^2) $ is

$$
    E(x) = \int\limits_{-\infty}^{+\infty} x \frac{1}{\sqrt{2\pi\sigma^2}}e^{\frac{(x)^2}{2\sigma^2}} = 0
$$

and variance is $ V(X) = \sigma^2 $
Since standard deviation $ \widehat \sigma $ equals $ \frac 1n \sum\limits\_{i=1}^n |X_i| $, then expectation of standard deviation is

$$
    E[\widehat \sigma] = E[\frac 1n \sum\limits_{i=1}^n |X_i|]
$$

According to the task $ [X_i] $ is sampled from the $ \mathcal{N}(0, \sigma^2) $ and thus, absolute value $ |X_i| $ follows half normal distribution. The expectation of half normal distribution is

$$
    E[ | X_i | ] = \frac{\sigma \sqrt{2}}{\sqrt{\pi}}
$$

Thus expectation of $ \widehat \sigma $ is

$$
    E[\widehat \sigma] = E[\frac{1}{n} \sum_{i=1}^n |X_i| ] = \frac{1}{n}\sum_{i=1}^n E[ |X_i| ] = \frac{1}{n}\sum_{i=1}^n \sigma \sqrt{\frac{2}{\pi}} = \sigma \sqrt{\frac{2}{n\pi}}
$$

Then bias is

$$
    Bias(\widehat \sigma) = E[\widehat \sigma] - \sigma = \sigma \sqrt{\frac{2}{n\pi}} - \sigma = \sigma (\sqrt{\frac{2}{n\pi}} - 1)
$$

$ \boldsymbol{(b)} $ In statistics, the standard error is the standard deviation of the sample distribution. General formula for standard error $ \boldsymbol{SE} $ is the square root of the variance

$$
    \boldsymbol{SE} = \sqrt{Var(\widehat \sigma)}
$$

Variance of the half normal distribution is

$$
    Var(| X_i |) = \sigma^2 (1 - \frac{2}{\pi})
$$

In our case

$$
    Var(\widehat \sigma) = Var(\frac{1}{n} \sum_{i=1}^n | X_i |) = \frac{1}{n^2} \sum^n_{i = 1} Var(|X_i|) = \frac{1}{n^2} n \sigma^2(1 - \frac{2}{\pi}) = \frac{\sigma^2}{n}(1 - \frac{2}{\pi})
$$

Then standart error $\boldsymbol{SE}$ is

$$
    \boldsymbol{SE(\widehat \sigma)} = \sqrt{Var(\widehat \sigma)} = \sqrt{\frac{\sigma^2}{n}(1 - \frac{2}{\pi})}
$$

$ \boldsymbol{(c)} $ The MSE is a measure of the overall accuracy of the estimator, and it is the sum of the squared bias and the variance

$$
    MSE(\widehat \sigma) = Bias(\widehat \sigma)^2 + Var(\widehat \sigma)
$$

By substituting Bias from (a) and Var from (b) we get

$$
    MSE(\widehat \sigma) = Bias(\widehat \sigma)^2 + Var(\widehat \sigma) = (\sigma (\sqrt{\frac{2}{n\pi}} - 1))^2 + \frac{\sigma^2}{n}(1 - \frac{2}{\pi}) = \sigma^2 (\sqrt{\frac{2}{n\pi}} - 1)^2 + \frac{\sigma^2}{n}(1 - \frac{2}{\pi})
$$
