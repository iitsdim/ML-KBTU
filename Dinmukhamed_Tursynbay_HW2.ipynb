## Task 1.1 (1 point)

Let $\xi \sim \mathrm{Bin}(n, p)$. Prove that $\mathbb E\xi = np$, $\mathbb V\xi = np(1-p)$.

### YOUR SOLUTION HERE

$E(X+Y) = E(X)+E(Y)$
for independent $X$ and $Y$

so 
$$
E(\xi) = E(1) + E(2) + ... + E(\xi) = \xi \cdot E(1) = \xi \cdot p
$$

$Var(X+Y) = VAR(X)+VAR(Y)$
for independent $X$ and $Y$

so 

$$
Var(\xi) = Var(1) + Var(2) + ... + Var(\xi) = \xi \cdot Var(1) = \xi \cdot p \cdot (1-p)
$$

## Task 1.3 (3 points)

Let $X_1, \ldots, X_n$ is an i.i.d. sample from $\mathcal N(0, \sigma^2)$ and $\widehat \sigma = \frac 1n \sum\limits_{i=1}^n |X_i|$.

* Find bias, se and MSE of the estimation $\widehat \sigma$. Is this estimation unbiased? 
* If $\widehat \sigma$ turned out to be biased, fix it to get an unbiased estimation of $\sigma$. Find se of this corrected estimation. Is it consistent?

### YOUR SOLUTION HERE

1.

bias of standard deviation $ \widehat \sigma $ is
$$
    bias(\widehat \sigma) = E[\widehat \sigma] - \sigma
$$
and for normal distribution $\mathcal N(0, \sigma^2)$
$$
    E(x) = \mu
$$

Since $\mu$ is $0$, $ E(x) = 0 $

$$
    E[\widehat \sigma] = E[\frac 1n \sum\limits_{i=1}^n |X_i|]
$$

$ X_i $ is sampled from the $ \mathcal{N}(0, \sigma^2) $ and therefore, absolute value $ |X_i| $ follows half normal distribution. The expectation of half normal distribution is

$$
    E[ | X_i | ] = \sigma \sqrt{\frac{2}{\pi}}
$$

Thus expectation of $ \widehat \sigma $ is

$$
    E[\widehat \sigma] = E[\frac{1}{n} \sum_{i=1}^n |X_i| ] = \frac{1}{n}\sum_{i=1}^n E[ |X_i| ] = \frac{1}{n}\sum_{i=1}^n \sigma \sqrt{\frac{2}{\pi}} = \sigma \sqrt{\frac{2}{n\pi}}
$$

Then bias is

$$
    Bias(\widehat \sigma) = E[\widehat \sigma] - \sigma = \sigma \sqrt{\frac{2}{n\pi}} - \sigma = \sigma (\sqrt{\frac{2}{n\pi}} - 1)
$$

2. 

Variance of the half normal distribution is

$$
    Var(| X_i |) = \sigma^2 (1 - \frac{2}{\pi})
$$

In our case

$$
    Var(\widehat \sigma) = Var(\frac{1}{n} \sum_{i=1}^n | X_i |) = \frac{1}{n^2} \sum^n_{i = 1} Var(|X_i|) = \frac{1}{n^2} n \sigma^2(1 - \frac{2}{\pi}) = \frac{\sigma^2}{n}(1 - \frac{2}{\pi})
$$

Then $\boldsymbol{SE}$ is

$$
    \boldsymbol{SE(\widehat \sigma)} = \sqrt{Var(\widehat \sigma)} = \sqrt{\frac{\sigma^2}{n}(1 - \frac{2}{\pi})}
$$


3.

The MSE is 

$$
    MSE(\widehat \sigma) = Bias(\widehat \sigma)^2 + Var(\widehat \sigma) = (\sigma (\sqrt{\frac{2}{n\pi}} - 1))^2 + \frac{\sigma^2}{n}(1 - \frac{2}{\pi}) = \sigma^2 (\sqrt{\frac{2}{n\pi}} - 1)^2 + \frac{\sigma^2}{n}(1 - \frac{2}{\pi})
$$

## Task 1.4 (3 points)

Let $X_1, \ldots, X_n$ is an i.i.d. sample form $U[0, 2\theta]$. The parameter $\theta$ can be estimated as
$\widehat\theta = \overline X_n$ (sample average) or $\tilde\theta = \mathrm{med}(X_1, \ldots, X_n)$ (sample median). According to ML Handbook, both estimations are unbiased, and $\mathbb V \widehat\theta = \frac{\theta^2}{3n}$, $\mathbb V \tilde\theta = \frac{\theta^2}{n+3}$. Hence, due to the central limit theorem

$$
    \widehat\theta \approx \mathcal N\Big(\theta, \frac{\theta^2}{3n}\Big), \quad
    \tilde\theta \approx \mathcal N\Big(\theta, \frac{\theta^2}{n+3}\Big), \quad n \gg 1.
$$

In this exercise you are suggested to verify this equalities numerically using NumPy, and plot the results (see example below).

### Possible result of the experiment

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_hists(theta, n):
    x = np.random.normal(theta, theta * theta / (n+3), n)
    y = np.random.normal(theta, theta * theta / (3*n), n)
    sns.histplot([x, y], color=['r', 'b'])
```
