```python
import numpy as np
from scipy.stats import binom, poisson, norm
```

Methods of [`scipy.stats.rv_discrete`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.rv_discrete.html#scipy.stats.rv_discrete):

* Probability mass function (pmf) `rv.pmf(k)`: $p_k = \mathbb P(\xi = k)$
* Cumulative distribution function (cdf) `rv.cdf(k)`: $\sum\limits_{i\leqslant k} p_i$
* Percent point function (ppf) `rv.ppf(q)`: $\min\limits_{k}\{\mathrm{cdf}(k) \geqslant q\}$
* Entropy `rv.entropy()`: $-\sum\limits_k p_k \log p_k$

Methods of [`scipy.stats.rv_continuous`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.rv_continuous.html#scipy.stats.rv_continuous):

* Probability density function (pdf) `rv.pdf(x)`: $p_\xi(x)$
* Cumulative distribution function (cdf) `rv.cdf(x)`: $F_\xi(x) = \mathbb P(\xi \leqslant x) = \int\limits_{-\infty}^x p_\xi(x)\,dx$
* Percent point function (ppf) `rv.ppf(q)`: $x\colon F_\xi(x) = q$
* Entropy `rv.entropy()`: $-\int\limits_{-\infty}^{+\infty} p_\xi(x) \log p_{\xi}(x)\,dx$

## Binomial distribution

```python
n, p = 6, 0.4
rv_binom = binom(n, p)
print("pmf:", rv_binom.pmf(np.arange(n + 1)))
print("cdf:", rv_binom.cdf(np.arange(n + 1)))
print("ppf:", rv_binom.ppf(0.2))
print("entropy:", rv_binom.entropy())
```

## Poisson distribution

```python
mu = 15
rv_poisson = poisson(mu)
print("pmf:", rv_poisson.pmf(np.arange(2*mu)))
print("cdf:", rv_poisson.cdf(np.arange(2*mu)))
print("ppf:", rv_poisson.ppf(0.8))
print("entropy:", rv_poisson.entropy())
```

## Gaussian distribution

```python
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'svg'

xs = np.linspace(-4, 4, 501)
plt.plot(xs, norm.pdf(xs), lw=2, c='r', label="pdf")
plt.plot(xs, norm.cdf(xs), lw=2, c='b', label="cdf")
plt.axhline(c='k', alpha=0.7)
plt.axvline(c='k', alpha=0.7)
plt.xlim(-4, 4)
plt.legend()
plt.grid(ls=":");
```
